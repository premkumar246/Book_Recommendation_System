# What is a Recommender system?

![Recommender System](https://www.shutterstock.com/image-vector/recommender-system-provides-smart-recommendations-260nw-1978170500.jpg)

A Recommender system is a system which suggest items like songs, books, movies or goods based on the preference of the user to make user more engaged in to the business and ultimately generate profits by increasing sales. 

There are 4 types of recommender systems:<br>
**1. Popularity Based Recommender System**<br>
**2. Content Based Recommender System**<br>
**3. Collaborative Filtering Based Recommender System**<br>
**4. Hybrid Recommender System**<br>
Lets look in to each type and understand what these means. 

#### **1. Popularity Based Recommender System:**
* This system takes the best items of all and put forth to the audiance so that they can get thier hands on what is popular or the best currently. 
* The system uses a formula which is decided by the company, based on the formula the system recommends.<br>

Example: 
   * YouTube - Trending videos 
   * IMDB - Top 250 Best Movies

#### **2. Content Based Recommender System**
* This system consider the attributes and based on the similarity of the attributes the system recommends the content which share similar values for the attributes like Genre, Movie name, Music director, Hero etc. <br>

Example:<br>
   * Music director wise playlist of songs in spotify
   * Genre based movies in Netflix

#### **3. Collaborative Filtering Based Recommender System**
* This system takes the ratings of users for movies plot on the feature space and try to cluster the similar movies based on similarity distance methods.
* Whatever the movies is very near to the movies watched by a person those movies are recommended to the user

Example:
   * Products recommended by amazon. 

#### **4. Hybrid Recommender System**
* This system will work by combining any two or all three above mentioned methods. 

Examples:
* Netflix uses Hybrid system for recomending the movies

# Popularity Based Recommender System

* Since we know popularity based recommender system is a system which recommends the best or most popular items of all based on a formula. 
* Here let us make a list of books which are popular among most and recommened them to the user. 
* We consider those books which got atleast 250 votes from the user as top 50 books. 
* For this we need books and ratings in same dataset. 

```python
# Let's get the books and ratings in to one dataframe. 
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name
```

```python
# Lets count number of votes got by each book
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
# Create a dataframe with number of ratings and average ratings columns
popularity_df= pd.concat([num_rating_df,avg_rating_df['avg_ratings']],axis=1)
popularity_df
```

```python
popular_50_books = popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(60)
popular_50_books
```

```python
popular_50_df = popular_50_books.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num_ratings','avg_ratings']]
popular_50_df
```

# Collaborative Filtering Recommender System

### Approach <br>
* We know collaborative filtering system recommends books based on similarity by clustering the books on any of the numerical variable. 
* Now to proceede for clustering we should decide on what variable we can cluster the books. 
* Let us cluster the books based on ratiings, which tells about how likely the book is for users. 
* To cluster the books based on ratings we need a dataframe whose data points can be represented in feature space and then using similarity techniques we apply clustering. 
* Now we will build a dataframe where users will be the columns and books will be the index and values will be the ratings provided by the users for each book. 
* But here is a problem not every user rates all books and not every book gets ratings. 
* there are users who rated 0 books and there are books who got 0 ratings. 
* If we consider this sort of data with null values clustering may not be effective and the recommender system will not work efficiently. 
* To overcome this we will choose user who have rated atleast 200 books and books with atleast 50 user ratings.
* This solves the problem of null values and the data we consider filters most reliable users and popular books. 

```python
# Lets filter user who rated atleast 200 books
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
pdhe_likhe_users = x[x].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(pdhe_likhe_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns ='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Zoya')
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1],reverse=True)[1:6]
    
    
    data=[]
    for i in similar_items:
        
        item=[]
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
